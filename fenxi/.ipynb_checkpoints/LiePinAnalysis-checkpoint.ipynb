```python
#导入数据
# coding: utf-8 
import pandas as pd
import pymysql
db = pymysql.connect(host="localhost",user="root",passwd="123456",db="LiePinData",charset = 'utf8')

sql = "select distinct jobtitle, company,min_salary,max_salary,average_salary,pdate,tag_list,\
description,industry,companysize,comaddress,position1,education,workexperience,worklanguage,age,workyear \
from `data`\
where (jobtitle LIKE '%数据%分析%' \
or jobtitle LIKE'%大数据%'\
or jobtitle like '%数据%挖掘%'\
OR jobtitle LIKE '%数据%运营%'\
OR jobtitle LIKE '%data%')\
and is_end=0 "
data = pd.read_sql(sql,db)
data['count'] = 1
#data['workyear']=float(data['workyear'])
db.close()
data
```

```python
#行业分布_直方图-麻烦的写法
from pyecharts import Bar
from pyecharts import Grid
grid = Grid()

industry_dis = data.groupby('industry')['count'].count()
#print(industry_dis)
industry_dict = {}
for i in range(0,len(industry_dis.index)):
    industry_dict[industry_dis.index[i]] = int(industry_dis[i])
#print(industry_dict)

dict = sorted(industry_dict.items(), key = lambda asd:asd[1],reverse = True)
print(dict)
industry = []
count = []
for item in dict:
    if item[1]>0:
        industry.append(item[0])
        count.append(item[1])
print(industry)
print(type(count[0]))
bar = Bar()
bar.add("", is_convert = False,is_yaxis_inverse = False,is_label_show = 1 ,\
        x_axis = industry,xaxis_interval = 0,xaxis_rotate = -45,xaxis_name = '',\
        y_axis= count,yaxis_name = '职位数量',yaxis_force_interval = 100,yaxis_name_gap = 40,\
       is_datazoom_show = True,datazoom_orient = 'horizontal',datazoom_type = 'both'\
       )
bar
```

```python
#行业分布-职位/公司-柱状图
from pyecharts import Bar

groupby_industry = data.groupby('industry').agg({'jobtitle':pd.Series.count,'company':pd.Series.nunique})\
.sort_values(['jobtitle', 'company'], ascending=[0,0 ])
industry = list(groupby_industry.index)
jobcount = list(map(int,groupby_industry['jobtitle']))
companycount = list(map(int,groupby_industry['company']))
#print(industry)
#print(type(jobcount[0]))
#print(type(companycount[0]))

bar = Bar()
bar.add("职位数量",x_axis = industry,xaxis_interval = 0,bar_category_gap=0,\
       y_axis = jobcount,is_label_show = 1,is_stack=0)
bar.add("公司数量", is_convert = False,is_yaxis_inverse = False,is_label_show = 1 ,\
        x_axis = industry,xaxis_interval = 0,xaxis_rotate = -45,xaxis_name = '',\
        y_axis= companycount,yaxis_name = '职位数量',yaxis_force_interval = 100,yaxis_name_gap = 40,\
       is_datazoom_show = True,datazoom_orient = 'horizontal',datazoom_type = 'both',is_stack=0,bar_category_gap=0\
       )

bar
```

```python
#行业分布_饼图
from pyecharts import Bar,Pie
ind = []
jcount = []
ccount = []
elsesum_j = 0
elsesum_c = 0
for i in range(0,len(industry)):
    if jobcount[i]>=80:
        ind.append(industry[i])
        jcount.append(jobcount[i])
        ccount.append(companycount[i])
    else:
        elsesum_j += jobcount[i]
        elsesum_c += companycount[i]
ind.append('其他')
jcount.append(elsesum_j)
ccount.append(elsesum_c)


pie = Pie()
pie.add('',ind,jcount,radius=[60,75],is_label_show = 1)
pie.add('',ind,ccount,radius = [0,20],\
        is_label_show =1,label_formatter = '{b}'+'\n'+'{d}%')
pie
```

```python
def get_salary_ind(grouped_Data,industry):
    result = []
    for item in grouped_Data:
        if item[0] == industry:
            result = list(item[1]['average_salary'])
    return result
```

```python
#行业-经验-箱形图
from pyecharts import Scatter,HeatMap,Boxplot
df = data[data['average_salary']!=0][['workyear','industry','average_salary']].groupby(['industry'])
ind = ['互联网·电商','投资','计算机软件','IT服务','银行']
y = []

for i in ind:
    salary = get_salary_ind(df,i)
    y.append(salary)
#print(y)

box = Boxplot()
yaxis = box.prepare_data(y) 
box.add('',ind,yaxis,xaxis_interval = 0,\
       is_data_zoom_show = True)
box
```

```python
def worky_salary_eachind(data,industry):
    df = data[(data['average_salary']!=0)&(data['industry']==industry)]\
    [['workyear','industry','average_salary']].groupby(['workyear']).agg({'average_salary':pd.Series.mean})
    ind = industry
    x_ind =[i for i in df.index]
    y_ind = [round(float(i),1) for i in list(df['average_salary'])]
    dic = {
        'ind':ind,
        'x':x_ind,
        'y':y_ind
    }
    return dic
```
